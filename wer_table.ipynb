```python
test_txt_files = []
test_audio_files = []
import os 

path_test = "/Users/aslonkhamidov/Downloads/ISSAI_USC/test/"
path_train = "/Users/aslonkhamidov/Downloads/ISSAI_USC/train/"
path_val = "/Users/aslonkhamidov/Downloads/ISSAI_USC/dev/"
files_test = os.listdir(path_test)
files_train = os.listdir(path_train)
files_val = os.listdir(path_val)
```

```python
ids_test = set([f.split(".")[0] for f in files_test])
ids_train = set([f.split(".")[0] for f in files_train])
ids_val = set([f.split(".")[0] for f in files_val])
```

```python
from dataclasses import dataclass

@dataclass
class AudioFile:
    id: str
    audio_path: str 
    txt_path: str
    text : str

    duration: float
```

```python
objects  = []
for idx in ids_test:
    txt_path = path_test + idx + ".txt"
    audio_path = path_test + idx + ".wav"
    with open(txt_path, "r") as f:
        text = f.read()
        objects.append(AudioFile(id=idx, audio_path=audio_path, txt_path=txt_path, text=text, duration=0.0))
```

```python
missing = 0
for idx in ids_train:
    txt_path = path_train + idx + ".txt"
    audio_path = path_train + idx + ".wav"
    try:
        with open(txt_path, "r") as f:
            text = f.read()
            objects.append(AudioFile(id=idx, audio_path=audio_path, txt_path=txt_path, text=text, duration=0.0))
    except:
        missing += 1
        continue
        
```

```python
missing = 0
for idx in ids_val:
    txt_path = path_val + idx + ".txt"
    audio_path = path_val + idx + ".wav"
    try:
        with open(txt_path, "r") as f:
            text = f.read()
            objects.append(AudioFile(id=idx, audio_path=audio_path, txt_path=txt_path, text=text, duration=0.0))
    except:
        missing += 1
        continue
```

```python
def calculate_duration(audio_path):
    import wave
    import contextlib
    with contextlib.closing(wave.open(audio_path,'r')) as f:
        frames = f.getnframes()
        rate = f.getframerate()
        duration = frames / float(rate)
        return duration
for o in objects:
    o.duration = calculate_duration(o.audio_path)
```
