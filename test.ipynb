```python
import requests
import json
```

```python
# Test cases for different weather scenarios
test_cases = [
    {
        "name": "Sunny day in Tashkent",
        "payload": {
            "backend_output": {
                "city": "Tashkent",
                "condition": "Sunny",
                "temperature": 35,
                "feels_like": 38,
                "humidity": 30,
                "sunrise": "5:45",
                "wind_speed": 5,
                "sunset": "19:30"
            },
            "function_name": "weather",
            "llm_output": "Here's the current weather in Tashkent."
        }
    },
    {
        "name": "Rainy day in Samarkand",
        "payload": {
            "backend_output": {
                "city": "Samarkand",
                "condition": "Rain",
                "temperature": 18,
                "feels_like": 16,
                "humidity": 85,
                "sunrise": "6:00",
                "wind_speed": 15,
                "sunset": "18:30"
            },
            "function_name": "weather",
            "llm_output": "It's currently raining in Samarkand."
        }
    },
    {
        "name": "Snowy day in Namangan",
        "payload": {
            "backend_output": {
                "city": "Namangan",
                "condition": "Snow",
                "temperature": -5,
                "feels_like": -10,
                "humidity": 90,
                "sunrise": "7:15",
                "wind_speed": 8,
                "sunset": "17:00"
            },
            "function_name": "weather",
            "llm_output": "It's snowing in Namangan today."
        }
    }
]

# Run tests and collect results
test_results = []
for test in test_cases:
    print(f"Running test: {test['name']}")
    resp = requests.get("http://localhost:8000/chat/v3/build_ui", json=test["payload"])
    test_results.append({
        "name": test["name"],
        "status_code": resp.status_code,
        "response": resp.json() if resp.status_code == 200 else resp.text
    })
    print(f"Status code: {resp.status_code}")
    print(f"Response: {resp.json() if resp.status_code == 200 else resp.text}")

# Save test results to file
with open("test_results.json", 'w') as f:
    json.dump(test_results, f, ensure_ascii=False, indent=4)

# Print summary
print("\nTest Summary:")
for result in test_results:
    status = "✅ PASS" if result["status_code"] == 200 else "❌ FAIL"
    print(f"{status} - {result['name']} (Status: {result['status_code']})")
```

```python
import pydivkit as dv
def _build_row(icon_url: str, label: str) -> dv.DivContainer:
    return dv.DivContainer(
        orientation=dv.DivContainerOrientation.HORIZONTAL,
        items=[
            dv.DivImage(
                image_url=icon_url,
                width=dv.DivFixedSize(unit=dv.DivSizeUnit.DP, value=24),
                height=dv.DivFixedSize(unit=dv.DivSizeUnit.DP, value=24)
            ),
            dv.DivText(
                text=label,
                font_size=16,
                text_color="#6F7580",
                paddings=dv.DivEdgeInsets(left=10)
            )
        ]
    )
card = dv.DivContainer(
    log_id="my_home_card",
    states=[
        dv.DivState(
            state_id=0,
            div=dv.DivContainer(
                orientation=dv.DivContainerOrientation.HORIZONTAL,
                width=dv.DivMatchParentSize(),
                height=dv.DivWrapContentSize(),
                items=[
                    # House Image
                    dv.DivImage(
                        image_url="https://example.com/house.png",  # Replace with a real hosted image URL
                        width=dv.DivFixedSize(unit=dv.DivSizeUnit.DP, value=160),
                        height=dv.DivFixedSize(unit=dv.DivSizeUnit.DP, value=160),
                        scale=dv.DivImageScale.FIT
                    ),
                    # Text and icons
                    dv.DivContainer(
                        orientation=dv.DivContainerOrientation.VERTICAL,
                        width=dv.DivMatchParentSize(),
                        height=dv.DivWrapContentSize(),
                        items=[
                            dv.DivText(
                                text="Мой дом",
                                font_size=24,
                                font_weight=dv.DivFontWeight.BOLD,
                                text_color="#1A1A33",
                                underline=dv.DivLineStyle.SINGLE
                            ),
                            dv.DivContainer(
                                orientation=dv.DivContainerOrientation.VERTICAL,
                                separator=dv.DivContainerSeparator(
                                    margins=dv.DivEdgeInsets(top=10, bottom=10)
                                    
                                    ),
                                items=[
                                    _build_row("https://example.com/lightbulb.png", "Электричество"),
                                    _build_row("https://example.com/fire.png", "Газ"),
                                    _build_row("https://example.com/wifi.png", "Интернет")
                                ]
                            )
                        ]
                    )
                ],
                paddings=dv.DivEdgeInsets(all=16),
                background=[
                    dv.DivBackground(
                        color="#FFFFFF",
                        corner_radius=dv.DivCornerRadius(all=24)
                    )
                ]
            )
        )
    ]
)

with open("card.json", "w") as f:
    json.dump(dv.make_div(card), f)

```
